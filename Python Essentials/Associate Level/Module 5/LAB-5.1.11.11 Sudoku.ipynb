#### Scenario
    As you probably know, Sudoku is a number-placing puzzle played on a 9x9 board. The player has to fill the board in a very specific way:
   - each row of the board must contain all digits from 0 to 9 (the order doesn't matter)
   - each column of the board must contain all digits from 0 to 9 (again, the order doesn't matter)
   - each of the nine 3x3 "tiles" (we will name them "sub-squares") of the table must contain all digits from 0 to 9.

#### Your task is to write a program which:
   - reads 9 rows of the Sudoku, each containing 9 digits (check carefully if the data entered are valid)
   - outputs Yes if the Sudoku is valid, and No otherwise.

```python
# Python3 program to check whether  given sudoku board is valid or not  
"""
    Step 1: Checks whether there is any duplicate in current row or not
    Step 2: Checks whether there is any duplicate in current column or not.
    Step 3: Checks whether there is any duplicate in current 3x3 box or not.  
    Step 4: Checks whether current row , current column and current 3x3 box is valid or not 
    Step 5: checks whether configuration is valid or not
"""
#Step 1   
def row_check(arr, row):  
  
    # set_char to store characters in the row.  
    set_char = set()  
  
    for i in range(0, 9):  
        
        # if the value already exists,return false  
        if arr[row][i] in set_char:  
            
            return False
  
        # if the current value in the row not empty, add the value in the set  
        if arr[row][i] != '.':  
            set_char.add(arr[row][i])  
    
    return True

# Step 2  
def col_check(arr, col):  
  
    set_char = set()  
  
    for i in range(0, 9):  
  
        # if the valuealready exists,return false
        if arr[i][col] in set_char: 
            return False
  
        # if the current value in the column is not empty, add the value in the set  
        if arr[i][col] != '.':  
            set_char.add(arr[i][col])  
    
    return True

# Step 3
def box_check(arr, startRow, startCol):  
  
    set_char = set()  
  
    for row in range(0, 3):  
        for col in range(0, 3):  
            curr_value = arr[row + startRow][col + startCol]  
  
            # if the value already exists, return false  
            if curr_value in set_char:  
                return False
  
            # if the current value is not empty, add the current value in the set  
            if curr_value != '.':  
                set_char.add(curr_value)  
         
    return True

# Step 4 
def isValid(arr, row, col):  
    
    return (row_check(arr, row) and col_check(arr, col) and
            box_check(arr, row - row % 3, col - col % 3))  
# Step 5
def isValidConfig(arr, n):  
  
    for i in range(0, n):  
        for j in range(0, n):  
  
            # If current row or column or 3x3 box is not valid, return false  
            if not isValid(arr, i, j):  
                return False
          
    return True  


# Drivers code  
if __name__ == "__main__": 
  
    board = [[ '5', '3', '.', '.', '7', '.', '.', '.', '.' ],  
             [ '6', '.', '.', '1', '9', '5', '.', '.', '.' ],  
             [ '.', '9', '8', '.', '.', '.', '.', '6', '.' ],  
             [ '8', '.', '.', '.', '6', '.', '.', '.', '3' ],  
             [ '4', '.', '.', '8', '.', '3', '.', '.', '1' ],  
             [ '7', '.', '.', '.', '2', '.', '.', '.', '6' ],  
             [ '.', '6', '.', '.', '.', '.', '2', '8', '.' ],  
             [ '.', '.', '.', '4', '1', '9', '.', '.', '5' ],  
             [ '.', '.', '.', '.', '8', '.', '.', '7', '9' ]] 
          
    if isValidConfig(board, 9):  
        print("YES") 
    else: 
        print("NO") 

```
