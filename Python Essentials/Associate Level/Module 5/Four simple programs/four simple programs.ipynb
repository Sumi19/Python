```python
def Caesar_cipher():
    msg = input("Enter your message: ")
    cipher=' '
    for char in msg:
        if not char.isalpha():
            continue
        char = char.upper()
        code = ord(char) + 1
        if code > ord('Z'):
            code = ord("A")
        cipher += chr(code)
    return cipher

if __name__ == '__main__':
    cipher = Caesar_cipher()
    print()
    print("your message ",cipher)
```

```python
def Caesar_cipher_decrypt():
    cipher = input("Enter your cipher: ")
    msg=' '
    for char in cipher:
        if not char.isalpha():
            continue
        char = char.upper()
        code = ord(char) - 1
        if code < ord('A'):
            code = ord("Z")
        msg += chr(code)
    return msg

if __name__ == '__main__':
    msg = Caesar_cipher_decrypt()
    print()
    print("your message ",msg)
```

```python
def num_processor():
    number = input("Enter a line of numbers - separate them with spaces: ")
    strings = number.split()
    total = 0
    try:
        for substr in strings:
            total += float(substr)
        return total
    except ValueError:
        print(substr,"is not a number")
        num_processor()

if __name__ == '__main__':
    
    x = num_processor()
    print()
    print('The total of the given number is :',x)
```
